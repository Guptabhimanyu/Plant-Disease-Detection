```python
import keras
import cv2
import numpy as np
from tensorflow.keras.preprocessing import image
import os
```

```python

# Load the trained model from Google Drive
model_path = 'C:/Users/Aayush Walmiki/Desktop/model2.h5'
model = keras.models.load_model(model_path)

# Define the categories for your classes
class_labels = ['Pepper,_bell___Bacterial_spot', 'Pepper,_bell___healthy', 'Potato___Early_blight', 'Potato___Late_blight', 'Potato___healthy'] # modify with your class names
```

```python
def predict_image(img):
    # Preprocess the image
    img = cv2.resize(img, (224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img/255.0
    
    # Make a prediction on the preprocessed image
    predictions = model.predict(img)
    
    # Get the predicted class index
    predicted_class_index = np.argmax(predictions, axis=-1)[0]
    
    # Get the predicted class label
    predicted_class_label = class_labels[predicted_class_index]
    
    # Get the probability of the predicted class
    predicted_class_prob = predictions[0][predicted_class_index]
    
    return predicted_class_label, predicted_class_prob

# Initialize the webcam
cap = cv2.VideoCapture(0)

# Define a loop to capture and predict on each frame from the webcam
while True:
    # Capture a frame from the webcam
    ret, frame = cap.read()
    
    # Make a prediction on the captured frame
    predicted_class_label, predicted_class_prob = predict_image(frame)
    #
    if predicted_class_prob > 0.90:
        text_pos = (10, 50)
    # Display the predicted class label on the frame
        cv2.putText(frame, f"{predicted_class_label}: {predicted_class_prob:.2f}",text_pos, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    
    # Display the frame
    cv2.imshow('Plant Disease Detection', frame)
    
    # Exit the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
